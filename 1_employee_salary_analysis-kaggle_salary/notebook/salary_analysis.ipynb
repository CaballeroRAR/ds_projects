## Distribution analysis


How are employees distributed across Departments, Education levels, Cities, and Gender.

```python
e_per_dept = data.groupby('Department', as_index=False).agg(
    Employee_Count=('EmployeeID', 'count'),
    Total_Salaries_Dept = ('Monthly_Salary', 'sum'),
)
e_per_dept['salary_dept_ratio'] = (e_per_dept['Total_Salaries_Dept'] / e_per_dept['Total_Salaries_Dept'].sum())
display(e_per_dept)
```

```python
# Create figure and axis
fig, ax1 = plt.subplots(figsize=(12, 7))

# X positions for bars
x = np.arange(len(e_per_dept))
width = 0.35

# Create bars for Employee Count (primary y-axis)
bars1 = ax1.bar(x - width/2, e_per_dept['Employee_Count'], width,
                label='Employee Count', color='steelblue', alpha=0.8)

# Set up primary y-axis (Employee Count)
ax1.set_xlabel('Department', fontsize=12, fontweight='bold')
ax1.set_ylabel('Employee Count', fontsize=12, fontweight='bold', color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')
ax1.set_xticks(x)
ax1.set_xticklabels(e_per_dept['Department'], rotation=0, fontsize=11)
ax1.set_title('Employee Count, Total Salaries and % of Total Salaries by Department',
              fontsize=14, fontweight='bold', pad=20)

# Set reasonable y-axis limits for Employee Count
max_employee = e_per_dept['Employee_Count'].max()
ax1.set_ylim(0, max_employee * 1.3)  # Add 30% padding

# Create secondary y-axis for Total Salaries
ax2 = ax1.twinx()

# Create bars for Total Salaries
bars2 = ax2.bar(x + width/2, e_per_dept['Total_Salaries_Dept'], width,
                label='Total Salaries ($)', color='coral', alpha=0.8)

# Set up secondary y-axis (Total Salaries)
ax2.set_ylabel('Total Salaries / Dept ($)', fontsize=12, fontweight='bold', color='coral')
ax2.tick_params(axis='y', labelcolor='coral')

# Format salary values on y-axis with commas and K/M/B suffixes for large numbers
def format_large_number(x, pos):
    if x >= 1e9:
        return f'${x/1e9:.1f}B'
    elif x >= 1e6:
        return f'${x/1e6:.1f}M'
    elif x >= 1e3:
        return f'${x/1e3:.1f}K'
    else:
        return f'${x:,.0f}'

ax2.yaxis.set_major_formatter(plt.FuncFormatter(format_large_number))

# Add value labels on top of bars
def add_value_labels(bars, axis, is_currency=False, is_large_number=False):
    for bar in bars:
        height = bar.get_height()
        if is_currency and is_large_number:
            if height >= 1e9:
                value = f'${height/1e9:.1f}B'
            elif height >= 1e6:
                value = f'${height/1e6:.1f}M'
            elif height >= 1e3:
                value = f'${height/1e3:.1f}K'
            else:
                value = f'${height:,.0f}'
        elif is_currency:
            value = f'${height:,.0f}'
        else:
            value = f'{height:,.0f}'
        axis.text(bar.get_x() + bar.get_width()/2, height * 1.02,
                 value, ha='center', va='bottom', fontsize=9,
                 fontweight='bold', color='black')

# Check if salary values are extremely large
if e_per_dept['Total_Salaries_Dept'].max() > 1e6:
    is_large_salary = True
else:
    is_large_salary = False

add_value_labels(bars1, ax1, is_currency=False, is_large_number=False)
add_value_labels(bars2, ax2, is_currency=True, is_large_number=is_large_salary)

# Calculate total salaries across all departments for percentage calculation
total_salaries_all_depts = e_per_dept['Total_Salaries_Dept'].sum()

# Add percentage values between the bars
for i in range(len(e_per_dept)):
    # Calculate percentage of total salaries
    dept_salary = e_per_dept['Total_Salaries_Dept'].iloc[i]
    salary_percentage = (dept_salary / total_salaries_all_depts) * 100

    # Position the percentage value in the center between the two bars
    ratio_x_position = x[i]  # Center of the department position

    # Get the heights of both bars
    employee_height = e_per_dept['Employee_Count'].iloc[i]
    salary_height = e_per_dept['Total_Salaries_Dept'].iloc[i]

    # Find which is taller (on their respective scales) for positioning
    # Normalize both to [0,1] range to compare relative heights
    employee_norm = employee_height / max_employee
    salary_max = e_per_dept['Total_Salaries_Dept'].max()
    salary_norm = salary_height / salary_max if salary_max > 0 else 0

    # Use the larger normalized height for positioning
    if employee_norm > salary_norm:
        display_height = employee_height * 1.15
    else:
        display_height = max_employee * 1.15

    # Add percentage text
    ax1.text(ratio_x_position, display_height,
             f'{salary_percentage:.1f}%',
             ha='center', va='bottom', fontsize=10,
             fontweight='bold', color='darkgreen',
             bbox=dict(boxstyle='round,pad=0.3',
                      facecolor='lightgreen', alpha=0.8))

# Add grid for better readability
ax1.grid(axis='y', alpha=0.3, linestyle='--')

# Set appropriate y-axis limits for ax2 based on data
if is_large_salary:
    # For large salaries, add more padding
    ax2.set_ylim(0, e_per_dept['Total_Salaries_Dept'].max() * 1.25)
else:
    ax2.set_ylim(0, e_per_dept['Total_Salaries_Dept'].max() * 1.15)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Add legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()

# Add a custom entry for the percentage in the legend
from matplotlib.patches import Patch
percentage_legend = Patch(facecolor='lightgreen', edgecolor='darkgreen',
                     label='% of Total Salaries', alpha=0.8)
ax1.legend(lines1 + lines2 + [percentage_legend],
           labels1 + labels2 + ['% of Total Salaries'],
           loc='upper left', framealpha=0.9)

plt.show()
```

## Data Cleaning

```python
# Selection of features
numeric_cols = ['Age', 'Monthly_Salary'] 
categoric_cols = ['Department', 'City', 'Gender']
ordinal_cols = ['Education_Level', 'Experience_Years'] 

# Copy dataframe
data_for_cluster = data.copy()
```

```python
# Encode ordinal columns
# Binning Experience_Years
exp_bins = [-1, 2, 5, 7, 10, float('inf')]  
exp_labels = [0, 1, 2, 3, 4] 
data_for_cluster['Experience_Years'] = pd.cut(data_for_cluster['Experience_Years'], bins=exp_bins, labels=exp_labels).astype(int)

# Encoding Education_Level
edu_order = ['High School', 'Bachelor', 'Master', 'PhD']
encoder = OrdinalEncoder(categories=[edu_order])
data_for_cluster['Education_Level'] = encoder.fit_transform(data_for_cluster[['Education_Level']])

data_for_cluster.head()
```

```python
# Encode categorical
data_for_cluster = pd.get_dummies(data_for_cluster, columns=categoric_cols, drop_first=True, dtype=int)
data_for_cluster.head()
```

```python
data_for_cluster.drop(columns=['Name'], inplace=True)
# Turn EmployeeID into index
data_for_cluster.set_index('EmployeeID', inplace=True)
data_for_cluster.info()
```

```python
# MAX-MIN scalation for numeric columns
colums_to_scale = [numeric_cols + ordinal_cols + categoric_cols]

data_for_cluster = data_for_cluster.astype(float) # To float all columns before scaling

scaler = MinMaxScaler() # All values are positive and whe need percentual interpretation
data_for_cluster[:] = scaler.fit_transform(data_for_cluster)

data_for_cluster.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA

# Apply PCA to reduce ALL features to 2D
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data_for_cluster)

# Transform centroids to the same PCA space
centroids_2d = pca.transform(centroids)

plt.figure(figsize=(12, 8))

# Create scatter plot
scatter = plt.scatter(data_2d[:, 0], data_2d[:, 1],
                     c=labels,
                     cmap=plt.cm.get_cmap('tab10', n_clusters),
                     s=60,
                     alpha=0.7,
                     edgecolors='w',
                     linewidth=0.5)

# Plot centroids
plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1],
           c='red',
           marker='X',
           s=300,
           label='Centroids',
           edgecolors='black',
           linewidth=2,
           zorder=10)

# Add centroid labels
for i, (x, y) in enumerate(centroids_2d):
    plt.text(x, y, f' C{i}', fontsize=12, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

# Calculate explained variance
explained_variance = pca.explained_variance_ratio_
total_variance = explained_variance.sum() * 100

plt.xlabel(f'Principal Component 1 ({explained_variance[0]*100:.1f}%)', fontsize=12)
plt.ylabel(f'Principal Component 2 ({explained_variance[1]*100:.1f}%)', fontsize=12)
plt.title(f'K-Means Clustering (k=4)\nPCA Projection of All Features\nTotal Variance Explained: {total_variance:.1f}%',
          fontsize=14, fontweight='bold')

plt.colorbar(scatter, label='Cluster ID')
plt.legend()
plt.grid(True, alpha=0.3, linestyle='--')
plt.tight_layout()
plt.show()

# Print PCA information
print(f"Original features: {data_encoded.shape[1]}")
print(f"Reduced to: 2 principal components")
print(f"Variance explained by PC1: {explained_variance[0]*100:.2f}%")
print(f"Variance explained by PC2: {explained_variance[1]*100:.2f}%")
print(f"Total variance explained: {total_variance:.2f}%")
```

```python
# Asegurarse de que las etiquetas de clster estn en el dataframe original (sin escalar)
if 'cluster' not in data.columns:
    data['cluster'] = labels  # labels son enteros

def describe_cluster(cluster_id):
    """Genera descripcin en lenguaje natural del cluster usando el dataframe original."""
    cluster_data = data[data['cluster'] == int(cluster_id)]
    if cluster_data.empty:
        return f"CLUSTER {cluster_id}: No hay datos para este cluster."

    # Modo de educacin (manejar caso vaco)
    edu_mode = cluster_data['Education_Level'].mode()
    edu_mode_str = edu_mode.iloc[0] if not edu_mode.empty else 'N/A'

    # Modo de departamento (manejar caso vaco)
    dept_mode = cluster_data['Department'].mode()
    dept_mode_str = dept_mode.iloc[0] if not dept_mode.empty else 'N/A'

    description = (
        f"CLUSTER {cluster_id} ({len(cluster_data)} personas, {len(cluster_data)/len(data):.1%}):\n"
        f" Edad promedio: {cluster_data['Age'].mean():.1f} aos\n"
        f" Salario promedio: ${cluster_data['Monthly_Salary'].mean():,.0f}\n"
        f" Experiencia promedio: {cluster_data['Experience_Years'].mean():.1f} aos\n"
        f" Educacin ms comn: {edu_mode_str}\n"
        f" Departamento ms comn: {dept_mode_str}\n"
    )

    return description

# Generar descripciones para todos los clusters (usar las etiquetas reales)
for i in sorted(np.unique(labels)):
    print(describe_cluster(i))
    print("-" * 50)
```
